Este fragmento de código tiene como objetivo identificar y contar cuántos valores 999 y 99 hay en cada columna numérica (int64 o float64) del DataFrame data. Estos valores (999 y 99) suelen usarse para representar datos faltantes o no disponibles en las encuestas o bases de datos.

```python
import numpy as np
import pandas as pd

def limpiar_variables_categoricas(data, excluir_columnas=None):
    if excluir_columnas is None:
        excluir_columnas = []

    # Verificar que data es un DataFrame
    if not isinstance(data, pd.DataFrame):
        raise TypeError("El parámetro 'data' debe ser un DataFrame de pandas.")

    # Verificar columnas
    columnas_validas = [col for col in data.columns if col not in excluir_columnas]

    columnas_categoricas = []
    for col in columnas_validas:
        try:
            if data[col].dtype.name in ['int64', 'float64', 'object']:
                columnas_categoricas.append(col)
        except Exception as e:
            print(f"Error al procesar columna {col}: {e}")

    data[columnas_categoricas] = data[columnas_categoricas].replace({999: 2, 99: np.nan})

    for col in columnas_categoricas:
        data[col] = data[col].astype('category')

    return data

```

# Ajustar a exportar  Excel

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency, fisher_exact
import statsmodels.api as sm

pd.set_option('display.multi_sparse', True)

def imprimir_resultados_chi2(data_prueba, columna_interes, variables_ind, exportar_excel=False, nombre_archivo="resultados_chi2.xlsx"):
    resultados = []

    variables_filtradas = [var for var in variables_ind if var in data_prueba.columns]

    if not variables_filtradas:
        print("⚠️ Ninguna de las variables en variables_ind está en el DataFrame.")
        return

    for columna in variables_filtradas:
        tabla_contingencia = pd.crosstab(data_prueba[columna], data_prueba[columna_interes])
        tabla_contingencia_1 = pd.crosstab(data_prueba[columna], data_prueba[columna_interes], margins=True, margins_name="Total")

        if tabla_contingencia.shape == (2, 2):
            # Chi2 sin corrección (Pearson puro)
            chi2_pearson, p_pearson, dof, _ = chi2_contingency(tabla_contingencia, correction=False)
            # Chi2 con corrección de continuidad (Yates)
            chi2_yates, p_yates, _, _ = chi2_contingency(tabla_contingencia, correction=True)

            n = tabla_contingencia.sum().sum()
            phi2 = chi2_pearson / n

            fila0, fila1 = tabla_contingencia.index
            col0, col1 = tabla_contingencia.columns

            prev_exp = tabla_contingencia.loc[fila1, col1] / tabla_contingencia.loc[fila1].sum()
            prev_no_exp = tabla_contingencia.loc[fila0, col1] / tabla_contingencia.loc[fila0].sum()
            porc_exp = tabla_contingencia.loc[fila1].sum() / n
            porc_no_exp = tabla_contingencia.loc[fila0].sum() / n

            print("\n---------------------------------------------------------------------------------")
            print(f"📌 Análisis de '{columna_interes}' vs '{columna}'")
            print("---------------------------------------------------------------------------------")
            print(f"📊 Tabla de Contingencia:\n\n{tabla_contingencia_1}\n")

            print(f"📈 Chi-cuadrado de Pearson: {chi2_pearson:.4f} | p-valor: {p_pearson:.4f}")
            print(f"📉 Chi-cuadrado con corrección de continuidad (Yates): {chi2_yates:.4f} | p-valor: {p_yates:.4f}")
            print(f"📏 Tamaño del efecto (Phi²): {phi2:.4f}")

            if (tabla_contingencia.values == 0).any():
                print("⚠️ Advertencia: La tabla contiene celdas con valor 0, Fisher puede no ser fiable.")
                fisher_or, fisher_p = np.nan, np.nan
            else:
                fisher_or, fisher_p = fisher_exact(tabla_contingencia)
                print(f"🎯 Prueba exacta de Fisher: OR = {fisher_or:.4f} | p-valor = {fisher_p:.4f}")

            table2x2 = sm.stats.Table2x2(tabla_contingencia.values)
            or_val = table2x2.oddsratio
            or_ci = table2x2.oddsratio_confint()
            rr_val = table2x2.riskratio
            rr_ci = table2x2.riskratio_confint()

            print(f"\n🔹 Odds Ratio (OR): {or_val:.4f}")
            print(f"📌 IC 95% OR: [{or_ci[0]:.4f}, {or_ci[1]:.4f}]")
            print(f"📌 Razón de Prevalencia (RP): {rr_val:.4f}")
            print(f"📌 IC 95% RP: [{rr_ci[0]:.4f}, {rr_ci[1]:.4f}]\n")

            print(f"📊 Prevalencia del evento en expuestos: {prev_exp:.2%}")
            print(f"📊 Prevalencia del evento en no expuestos: {prev_no_exp:.2%}")
            print(f"📊 Porcentaje de expuestos en la muestra: {porc_exp:.2%}")
            print(f"📊 Porcentaje de no expuestos en la muestra: {porc_no_exp:.2%}\n")

            resultados.append({
                 "Variable_dependiente": columna_interes,
                "Variable": columna,
                "Chi2_Pearson": chi2_pearson,
                "p_valor_Chi2_Pearson": p_pearson,
                "Chi2_Yates": chi2_yates,
                "p_valor_Yates": p_yates,
                "GL": dof,
                "Phi2": phi2,
                "Fisher_OR": fisher_or,
                "Fisher_p": fisher_p,
                "OR": or_val,
                "OR_IC_95_inf": or_ci[0],
                "OR_IC_95_sup": or_ci[1],
                "RP": rr_val,
                "RP_IC_95_inf": rr_ci[0],
                "RP_IC_95_sup": rr_ci[1],
                "Prev_evento_expuestos": prev_exp,
                "Prev_evento_no_expuestos": prev_no_exp,
                "%_expuestos": porc_exp,
                "%_no_expuestos": porc_no_exp
            })

        else:
            print(f"⚠️ La tabla entre '{columna}' y '{columna_interes}' no es 2x2. Se omite.\n")

    if exportar_excel and resultados:
        df_resultados = pd.DataFrame(resultados)
        df_resultados.to_excel(nombre_archivo, index=False)
        print(f"📁 Resultados exportados a: {nombre_archivo}")
```

```python
# Llamada a la función con nombres específicos
imprimir_resultados_chi2(
    data_prueba=data,
    columna_interes='P15_Tos_Ultimo_Año_Dif_Gripa',
    variables_ind=var_ind,
    exportar_excel=True,
    nombre_archivo="resultados_chi2.xlsx"
)
```
