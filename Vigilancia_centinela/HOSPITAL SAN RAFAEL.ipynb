```python
# Función para filtrar datos con solo ingresar el número de la semana
import pandas as pd

def filtrar_por_semana(df, semana):
    if semana not in semanas_2025:
        raise ValueError("Número de semana no válido. Debe estar entre 1 y 53.")

    Fecha_Inicio_Semana, Fecha_Final_Semana = semanas_2025[semana]

    # Convertir la columna de fechas a formato datetime si no lo está
    df["Fecha Ingreso"] = pd.to_datetime(df["Fecha Ingreso"])

    # Filtrar el DataFrame por la semana seleccionada
    df = df[
        (df["Fecha Ingreso"] >= Fecha_Inicio_Semana) &
        (df["Fecha Ingreso"] <= Fecha_Final_Semana)
    ]

    return df
```

```python
import pandas as pd

def reemplazar_ingreso_por_egreso(row):
    ingreso = str(row['Codigo CIE10']).strip().upper()
    egreso = str(row['CIE10 Egreso']).strip().upper()

    # Si son iguales, no hacer nada
    if ingreso == egreso:
        return row

     # Si ingreso empieza por I o J y egreso por otra letra → no se hace nada
    if ingreso.startswith(('I', 'J')) and not egreso.startswith(('I', 'J')):
        return row


    # Si ingreso no empieza por I o J y egreso sí → se reemplaza ingreso
    if not ingreso.startswith(('I', 'J')) and egreso.startswith(('I', 'J')):
        row['Codigo CIE10'] = egreso
        row["Diagnostico principal de Ingreso"] = row['Diagnostico Egreso']

    # En los demás casos, no se hace nada
    return row
```



####	Edad en meses: Calcular en caso de que la edad en años sea cero 0.
#### Edad años: Calcular de acuerdo a las variables fecha de nacimiento y fecha consulta.


```python
import pandas as pd
from dateutil.relativedelta import relativedelta

# Convertir las columnas a tipo datetime
df_sin_duplicados_columnas["Fecha de nacimiento"] = pd.to_datetime(df_sin_duplicados_columnas['Fecha de nacimiento'], format='%d/%m/%Y')
df_sin_duplicados_columnas['Fecha consulta'] = pd.to_datetime(df_sin_duplicados_columnas['Fecha consulta'], format='%d/%m/%Y')

# Función para calcular años y meses
def calcular_edad(row):
    nacimiento = row['Fecha de nacimiento']
    consulta = row['Fecha consulta']

    # Calcular la diferencia con relativedelta (maneja días/meses/años)
    delta = relativedelta(consulta, nacimiento)

    # Si hay al menos 1 año completo
    if delta.years >= 1:
        return pd.Series([delta.years, np.nan])
    else:
        # Calcular meses totales (años * 12 + meses)
        meses_totales = delta.years * 12 + delta.months
        return pd.Series([0, meses_totales])

# Aplicar la función y crear las columnas
df_sin_duplicados_columnas[['Edad años', 'Edad en meses']] = df_sin_duplicados_columnas.apply(calcular_edad, axis=1)
```

```python
df_sin_duplicados_columnas["Fecha de nacimiento"]=pd.to_datetime(df_sin_duplicados_columnas['Fecha de nacimiento']).dt.strftime('%d/%m/%Y')
df_sin_duplicados_columnas['Fecha consulta'] = pd.to_datetime(df_sin_duplicados_columnas['Fecha consulta']).dt.strftime('%d/%m/%Y')
```

### GRUPO DE EDAD: De acuerdo al intervalo
- 0 a 5 años: 1. INFANCIA

- 12 a 17 años: 3. ADOLESCENCIA

```python

# Función para asignar el grupo de edad
def clasificar_grupo_edad(edad):
    if 0 <= edad <= 5:
        return "1. INFANCIA"
    elif 6 <= edad <= 11:
        return "2. INFANCIA"
    elif 12 <= edad <= 17:
        return "3. ADOLESCENCIA"
    elif 18 <= edad <= 28:
        return "4. JUVENTUD"
    elif 29 <= edad <= 59:
        return "5. ADULTEZ"
    else:
        return "6. VEJEZ"


```
